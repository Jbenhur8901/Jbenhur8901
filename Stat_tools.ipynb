```python
from matplotlib import pyplot as plt
import pandas as pd 
from math import sqrt

"""This module contains useful statisticals methods to work with data.
   The module is made up of 3 branches|| the first branch is about methods to compute the Measures of position
   ||The second branch is about methods to compute the Measures of spread || The last  branch groups usefuls methods 
   to compute values such as the cumulative frequency, relative frequency,..."""


class Quantitative:

  """Quantative is The parent class for all the Discrete and Continuous Variable that we'll study|
  it define the values of the variable of study, the number of observations and the relative frequency if given"""

  def __init__(self,iterable1=None,iterable2=None):
    self.char =iterable1
    self.num_obs = iterable2
    

class Discrete(Quantitative):

  """Discrete is the child class that inherit from Quantitative the following attribute:
  self.char--->the values of the variable||self.num_obs---> the number of observations ||
  """

  def __init__(self,iterable1=None, iterable2=None):
    super().__init__(iterable1,iterable2)

# I. The Measures of Position   

  def mode(self):

    """ The mode is the character associated to the highest number of observations ||
    compute two attributes || self.char = character|| self.num_obs = number of observation"""

    highest_obs = max(self.num_obs) # find the highest number of observations

    for i,j in zip(self.char,self.num_obs):
        if j==highest_obs in (i,j): # check the couple (character,observation), return only the character
            mde = i
    return mde

  def ar_mean(self):
    """the mean is the average character of a sample of data||compute two attributes ||
     self.char = character|| self.num_obs = number of observation"""

    product =[]

    for i,j in zip(self.char,self.num_obs):
        product.append(i*j) # compute xi times ni
    mean = sum(product)/sum(self.num_obs)
    return round(mean,2)

  @staticmethod
  def quantiles(interable,rank):

    """The quantiles are the measures of position that separate the sample into several part
        like 25%,50% and 75% |takes an iterable and a float as arguments ||
        iterable = character|| rank = [0,25, 0,50, 0,75] """

    sequence = interable
    rank = rank
    n = len(sequence) # n refers to the total number of obersrvations
    n_rank = n*rank 
    sequence.sort()

    # if n_rank is an integer --> n must be even. Else, n is odd
    if type(n_rank)=="int": 
      quart = (sequence[n_rank]+sequence[n_rank+1])/2
    else:  
      quart = sequence[round(n_rank)+1]

    return quart

  # II. The Measures of Spread

  def s_range(self):

    """The range is the difference between maximun value and the minimum of a character 
    || works on self.char"""

    rng = max(self.char)-min(self.char)
    return round(rng,3)


  def var1(self):

    """return the weighted variance of a statistic serie|| takes as argument two interable
    """

    mean = self.ar_mean()
    numerator = []
    for i,j in zip(self.char, self.num_obs):
        product = ((i-mean)**2)*j # compute ni times the square of xi
        numerator.append(product)
    vrc1 = (sum(numerator)/sum(self.num_obs))
    return round(vrc1,2)

  def var2(self):

    """return the weighted variance of a statistic serie||"""

    mean = self.ar_mean()
    numerator = []
    for i,j in zip(self.char, self.rel_freq()):
        product = ((i-mean)**2)*j # compute ni times the square of xi
        numerator.append(product)
    vrc2 = sum(numerator)
    return round(vrc2,2)   

  def st_dev(self):

   """Compute the standard deviation of statistical serie"""
   sdv = sqrt(self.var2())
   return round(sdv,2)

  #Usefuls tools 
  def cum_freq1(self):

    """return the ascending cumulative frequency of a statistic serie in form of a list"""

    list_cf = [self.num_obs[0]] # withdraw the first value the frequency and add to the list of cum frequency "list_cf"
    index = 0
    for i in self.num_obs[1:]:
      add = list_cf[index] + i # add the previous element to the current value of i
      list_cf.append(add)
      index+=1 # increase the value of index # loop over the frequency
           
    return list_cf

  def cum_freq2(self):
    """return the cumulative relative frequency of a statistic serie in form of a list"""
    
    frequency = self.rel_freq()
        
    list_cf = [frequency[0]] # withdraw the first value the reverse frequency and add to the list of cum frequency "list_cf"
    index = 0
    for i in frequency[1:]:
      add = list_cf[index] + i # add the previous element to the current value of i
      list_cf.append(add)
      index+=1 # increase the value of index # loop over the frequency

    return list_cf

  @staticmethod
  def class_center(interable):
    """return the centers of all the classes of a statistic serie || takes as argument interable = intervals"""
    
    interval = list(interable)
    center = []
    for i in interval:
        result = sum(i)/2
        center.append(result)
    return center

  def rel_freq(self):
    """return the relative frequency of a statistic serie || takes as argument interable = frequency"""

    rel_fr = []
    for i in self.num_obs:
        calcul = i/sum(self.num_obs)
        rel_fr.append(calcul)
    return rel_fr

  @staticmethod
  def occurence(interable):
    #Return the occurence of specifics values of an interable in the form of a dictionnary ||"value": occurence
    list1 = [i for i in interable]
    dict1 = dict()
    for i in list1:
        x= list1.count(i)
        dict1[i]=x
    return dict1
    

class Continuous(Discrete):

  """Continuous inherit from Discrete all the method that can be shared between Discrete and Continuous||
  Process numerical dataset into continous statistical variable || contains methods for computing the
  number of classes needed to study a continuous variable, the amplitude of these classes and 
  finally set the classes in [a,b] form || if called as Get_interval(interable).myclasses(): return a 
  dictionnary in {class: frequency} form"""

  def __init__(self,iterable=None):
    
    self.total = len(iterable) # refers to the total of observations
    self.dict_serie = dict() # create the dictionnary that will contain data in form of {class: frequency}
    self.data = iterable # refers to the dataset
    self.class_num()   # return the number of classes needed for a given dataset
    self.class_length() # return the amplitude of the classes 
    self.interval(min(self.data)) # set class in form of [a,b] classes

    # list of all the class limits
    self.class_limit = {j[i] for j in Continuous.to_list(self.dict_serie.keys()) for i in range(2)}
    #Inherited attributes
    super().__init__(iterable1=[i for i in self.center(self.dict_serie.keys())],iterable2=[j for j in self.dict_serie.values()])
    
  
  def class_num(self):
    # apply the method of yule to get a number of classes 
    self.yule = 2.5*(self.total**0.25)
    return round(self.yule)
  
  def class_length(self):
    # return the amplitude of the class
    self.length = (max(self.data)-min(self.data))/self.yule
    return round(self.length,2)
    
  def interval(self,start,times=0):
    # return the classes and the number of observations for each class
    count = times
    r_freq = [] # list of the occurences of each value in the class
    

    # as long as count is less than or equal to the number of classes --> apply the recursion
    if count<= self.yule:
      count+=1 # increase count for each iteration
      end=round(start+self.length,2)
      for i in self.data:
        if start<=i<end: # get the value within the intervals
          cnt = self.data.count(i)
          r_freq.append(cnt)
          self.clas = [start,end]
    

      self.dict_serie[f"{self.clas}"] = len(r_freq) 
      self.interval(end,count)
  
  def myclasses(self):
    return self.dict_serie

  @staticmethod
  def center(iterable):
    convert =[]
    for i in iterable:
      x = [float(i[1:5]),float(i[7:11])]
      convert.append(x)
    interval = convert
    mid = []
    for i in interval:
      result = sum(i)/2
      mid.append(round(result,3))
    return mid 

  @staticmethod
  def to_list(iterable):

    convert =[]
    for i in iterable:
      x = [float(i[1:5]),float(i[7:11])]
      convert.append(x)

    return convert
   
  def infos(self):

    print(f"""
    the statistical serie: {self.dict_serie}
    the classes of this serie: {self.dict_serie.keys()}
    the correspondant number of observation: {self.dict_serie.values()}
    the total number of obervation: {self.total}
    the length of each class: {round(self.length,2)}
    the required number of classes: {round(self.yule)}
    the mean : {self.ar_mean()}
    the modal class: {self.mode()}
    the range: {self.s_range()}
    the variance : {self.var2()}
    the standard deviation: {self.st_dev()}
          """)
  
```
